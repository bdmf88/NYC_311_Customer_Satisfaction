```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Carregamento dos Dados ---

# O caminho para o seu arquivo local
caminho_arquivo = r"C:\Users\arthu\OneDrive\Documentos\DATA ANALYSIS\CSAT_311_Customer_Satisfaction_Survey\311_Customer_Satisfaction_Survey_Clean.xlsx"

# Carregando a aba específica "CLEAN" do arquivo Excel
try:
    df = pd.read_excel(caminho_arquivo, sheet_name="CLEAN")
    print("Arquivo carregado com sucesso!")
except FileNotFoundError:
    print(f"Erro: O arquivo não foi encontrado no caminho especificado:\n{caminho_arquivo}")
except Exception as e:
    print(f"Ocorreu um erro ao tentar ler o arquivo: {e}")

# --- Primeiros Passos da Análise Exploratória ---

# Verificando se o dataframe foi carregado antes de prosseguir
if 'df' in locals():
    # 1. Visualizar as 5 primeiras linhas para entender a estrutura
    print("\nAs 5 primeiras linhas do dataset:")
    display(df.head())

    # 2. Obter informações sobre os tipos de dados e valores nulos
    print("\nInformações sobre os tipos de dados e valores nulos:")
    df.info()

    # 3. Gerar estatísticas descritivas para colunas numéricas (como 'Year' e 'NPS')
    print("\nEstatísticas descritivas das colunas numéricas:")
    display(df.describe())
```

# Análise 1: Distribuição das Principais Métricas de Satisfação

```python
# --- Análise 1: Código (Versão Corrigida) ---

# --- 1.1) Criando a nova coluna 'NPS Category' ---

# Função para classificar a nota NPS
def classificar_nps(nota):
    if nota <= 6:
        return 'Detrator'
    elif nota <= 8:
        return 'Passivo'
    else:
        return 'Promotor'

# Aplicando a função para criar a nova coluna
df['NPS Category'] = df['NPS'].apply(classificar_nps)

print("Coluna 'NPS Category' criada com sucesso. Verificando as 5 primeiras linhas:")
# Mostra o resultado com a nova coluna
display(df[['NPS', 'NPS Category']].head())


# --- 1.2) Visualizando as distribuições ---

# Configurando o estilo e tamanho da figura
sns.set_style("whitegrid")
plt.figure(figsize=(18, 7))
plt.suptitle('Análise 1: Distribuição da Satisfação e Categorias NPS', fontsize=16, y=1.02)

# Gráfico 1: Distribuição da Satisfação Geral
plt.subplot(1, 2, 1)
ordem_satisfacao = ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree']
sns.countplot(data=df, x='Overall Satisfaction', order=ordem_satisfacao, palette='viridis')
plt.title('1.2.1) Distribuição da Satisfação Geral', fontsize=14)
plt.xlabel('Nível de Satisfação', fontsize=12)
plt.ylabel('Contagem de Respostas', fontsize=12)
plt.xticks(rotation=45)

# Gráfico 2: Distribuição das Categorias NPS
plt.subplot(1, 2, 2)
ordem_nps = ['Detrator', 'Passivo', 'Promotor']
sns.countplot(data=df, x='NPS Category', order=ordem_nps, palette='crest')
plt.title('1.2.2) Distribuição das Categorias NPS', fontsize=14)
plt.xlabel('Categoria NPS', fontsize=12)
plt.ylabel('Contagem de Respostas', fontsize=12)

# Ajusta e exibe os gráficos
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# Análise 2: Cálculo do Score NPS e Análise Temporal

```python
# --- Análise 2: Código (COM LEGENDA CORRIGIDA) ---

# --- 2.1) Cálculo do Score NPS Final ---
# (Esta parte do código não muda e pode ser mantida como estava)

# Contando o número de respostas em cada categoria
contagem_nps = df['NPS Category'].value_counts()
promotores = contagem_nps.get('Promotor', 0)
passivos = contagem_nps.get('Passivo', 0)
detratores = contagem_nps.get('Detrator', 0)
total_respostas = promotores + passivos + detratores
percent_promotores = (promotores / total_respostas) * 100
percent_detratores = (detratores / total_respostas) * 100
nps_score = percent_promotores - percent_detratores

print("--- 2.1) Cálculo do Score NPS Final ---")
print(f"Total de Respostas: {total_respostas}")
print(f"Promotores: {promotores} ({percent_promotores:.2f}%)")
print(f"Detratores: {detratores} ({percent_detratores:.2f}%)")
print(f"\nO Score NPS final é: {nps_score:.2f}")
print("-----------------------------------------\n")


# --- 2.2) Visualizando a Evolução do NPS por Ano (COM LEGENDA CORRIGIDA) ---

plt.figure(figsize=(13, 8)) # Aumentei um pouco a figura para dar espaço à legenda

# O seaborn já cria a legenda automaticamente ao usar o parâmetro 'hue'.
# A versão anterior pode ter tido um conflito no seu ambiente.
# Esta versão garante que a legenda seja posicionada corretamente.
ax = sns.histplot(data=df, x='Year', hue='NPS Category', multiple='fill', palette='viridis',
                  hue_order=['Detrator', 'Passivo', 'Promotor'], shrink=.8)

# Movendo a legenda para uma posição melhor (fora do gráfico)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1), title='Categoria NPS')

plt.title('2.2) Proporção de Categorias NPS por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Proporção de Respostas', fontsize=12)
plt.gca().yaxis.set_major_formatter('{:.0%}'.format) # Formata o eixo Y para porcentagem

plt.tight_layout() # Ajusta o layout para garantir que tudo fique visível
plt.show()
```

 # Análise 3 - Identificando os Fatores de Satisfação e Insatisfação

### 3.1) Relação entre o Conhecimento do Agente e o NPS

```python
# --- Análise 3: Código ---

# Criando o crosstab para calcular a contagem de cada combinação
ct = pd.crosstab(df['NPS Category'], df['Agent - Job Knowledge'])

# Calculando a proporção (porcentagem) em relação a cada Categoria NPS
ct_percent = ct.apply(lambda x: x / x.sum() * 100, axis=1)

# Reordenando as colunas para uma visualização lógica
ordem_satisfacao = ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree']
ct_percent = ct_percent[ordem_satisfacao]

print("--- 3.1) Tabela de Proporções: NPS Category vs. Agent - Job Knowledge ---")
display(ct_percent.round(2))
print("-" * 70)


# --- Visualização ---
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(14, 8),
                     colormap='coolwarm_r', rot=0)

plt.title('3.1) Impacto do Conhecimento do Agente na Categoria NPS', fontsize=16)
plt.xlabel('Categoria NPS', fontsize=12)
plt.ylabel('Proporção de Respostas (%)', fontsize=12)

# Colocando a legenda fora do gráfico
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1), title='Agent Job Knowledge')

# Adicionando os rótulos de porcentagem nas barras
for c in ax.containers:
    labels = [f'{w:.1f}%' if (w := v.get_height()) > 2 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='center', color='white', fontsize=10, fontweight='bold')


plt.tight_layout(rect=[0, 0, 0.85, 1]) # Ajusta o layout para a legenda
plt.show()
```

# Análise 4: Desempenho por Campanha de Pesquisa

```python
# --- Análise 4: Código ---

# Agrupando os dados por 'Campaign' e 'NPS Category' para contar as ocorrências
campaign_nps_counts = df.groupby(['Campaign', 'NPS Category']).size().unstack(fill_value=0)

# Verificando se todas as colunas necessárias existem
for col in ['Promotor', 'Passivo', 'Detrator']:
    if col not in campaign_nps_counts.columns:
        campaign_nps_counts[col] = 0

# Calculando o total de respostas por campanha
campaign_nps_counts['Total'] = campaign_nps_counts.sum(axis=1)

# Calculando as porcentagens e o NPS Score para cada campanha
campaign_nps_counts['% Promotor'] = (campaign_nps_counts['Promotor'] / campaign_nps_counts['Total']) * 100
campaign_nps_counts['% Detrator'] = (campaign_nps_counts['Detrator'] / campaign_nps_counts['Total']) * 100
campaign_nps_counts['NPS Score'] = campaign_nps_counts['% Promotor'] - campaign_nps_counts['% Detrator']

# Ordenando pelo NPS Score
campaign_nps_scores = campaign_nps_counts[['NPS Score']].sort_values(by='NPS Score', ascending=False)

print("--- 4.1) Tabela de Score NPS por Campanha ---")
display(campaign_nps_scores.round(2))
print("-" * 70)


# --- 4.2) Visualização do Score NPS por Campanha ---

plt.figure(figsize=(14, 8))
ax = sns.barplot(x=campaign_nps_scores.index, y=campaign_nps_scores['NPS Score'], palette='mako')

plt.title('4.2) Comparativo do Score NPS por Campanha', fontsize=16)
plt.xlabel('Campanha', fontsize=12)
plt.ylabel('NPS Score', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Adicionando os rótulos de dados
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points',
                fontsize=11)

plt.tight_layout()
plt.show()
```

# Análise 5: Análise de Satisfação por Idioma da Pesquisa

### Análise 5.1) Comparativo do Score NPS por Idioma

```python
# --- Análise 5.A: Código (NPS por Idioma) ---

print("--- Análise 5.A: NPS por Idioma ---")

# Contagem de respostas por idioma para referência
contagem_idiomas = df['Survey Language'].value_counts()
print("Contagem de Respostas por Idioma:")
display(contagem_idiomas)
print("-" * 70)

# Agrupando os dados por 'Survey Language' e 'NPS Category'
language_nps_counts = df.groupby(['Survey Language', 'NPS Category']).size().unstack(fill_value=0)

for col in ['Promotor', 'Passivo', 'Detrator']:
    if col not in language_nps_counts.columns:
        language_nps_counts[col] = 0

language_nps_counts['Total'] = language_nps_counts.sum(axis=1)
idiomas_relevantes = contagem_idiomas[contagem_idiomas > 30].index
language_nps_counts = language_nps_counts[language_nps_counts.index.isin(idiomas_relevantes)]

language_nps_counts['% Promotor'] = (language_nps_counts['Promotor'] / language_nps_counts['Total']) * 100
language_nps_counts['% Detrator'] = (language_nps_counts['Detrator'] / language_nps_counts['Total']) * 100
language_nps_counts['NPS Score'] = language_nps_counts['% Promotor'] - language_nps_counts['% Detrator']
language_nps_scores = language_nps_counts[['NPS Score']].sort_values(by='NPS Score', ascending=False)

print("\nTabela de Score NPS por Idioma (com mais de 30 respostas):")
display(language_nps_scores.round(2))
print("-" * 70)

# --- Visualização ---
plt.figure(figsize=(14, 8))
ax_nps = sns.barplot(x=language_nps_scores.index, y=language_nps_scores['NPS Score'], palette='viridis')
plt.title('5.A) Comparativo do Score NPS por Idioma', fontsize=16)
plt.xlabel('Idioma da Pesquisa', fontsize=12)
plt.ylabel('NPS Score', fontsize=12)

for p in ax_nps.patches:
    ax_nps.annotate(f'{p.get_height():.1f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 9), textcoords='offset points', fontsize=11)

plt.tight_layout()
plt.show()
```

### Análise 5.2) Análise das Métricas de CSAT por Idioma

```python
# --- Análise 5.B: Código (CSAT por Idioma) ---

print("\n--- Análise 5.B: Média de Satisfação (CSAT) por Idioma ---")

# Mapeando as respostas de satisfação para uma escala numérica
satisfaction_map = {
    'Strongly Disagree': 1,
    'Disagree': 2,
    'Neutral': 3,
    'Agree': 4,
    'Strongly Agree': 5
}
df['Overall Satisfaction Score'] = df['Overall Satisfaction'].map(satisfaction_map)

# Agrupando por idioma e calculando a média e a contagem
csat_por_idioma = df.groupby('Survey Language')['Overall Satisfaction Score'].agg(['mean', 'count'])

# Filtrando por idiomas relevantes (mesmo critério da análise NPS)
csat_por_idioma_relevante = csat_por_idioma[csat_por_idioma['count'] > 30].sort_values(by='mean', ascending=False)

print("\nTabela de Média de Satisfação por Idioma (escala de 1 a 5):")
display(csat_por_idioma_relevante.round(2))
print("-" * 70)

# --- Visualização ---
plt.figure(figsize=(14, 8))
ax_csat = sns.barplot(x=csat_por_idioma_relevante.index, y=csat_por_idioma_relevante['mean'], palette='plasma')
plt.title('5.B) Média de Satisfação Geral (CSAT) por Idioma', fontsize=16)
plt.xlabel('Idioma da Pesquisa', fontsize=12)
plt.ylabel('Média de Satisfação (de 1 a 5)', fontsize=12)
plt.ylim(0, 5) # Definindo o limite do eixo Y para a escala de 1 a 5

for p in ax_csat.patches:
    ax_csat.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 9), textcoords='offset points', fontsize=11)

plt.tight_layout()
plt.show()
```
